```python
import os
import pandas as pd

folder_path = 'data'

csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

dataframes = {}

for file in csv_files:
    file_path = os.path.join(folder_path, file)
    
    df = pd.read_csv(file_path)
    
    file_name = os.path.splitext(file)[0]
    dataframes[file_name] = df
```

```python
total_pop = dataframes['total population'][['GEO_ID','B01003_001E']].copy()
total_pop.columns = total_pop.iloc[0,:]
total_pop.drop(total_pop.index[0],inplace=True)
total_pop.columns.values[1] = 'total pop'
total_pop['total pop scaled'] = pd.to_numeric(total_pop['total pop'],errors='coerce') # since only standardization will be z-score, will write as scaled and others aswell
total_pop['total pop'].isnull().any()
total_pop.isnull().sum()
total_pop.sort_values(by='Geography',inplace=True)
```

```python
# DP02_0014E stimate!!HOUSEHOLDS BY TYPE!!Total households!!Households with one or more people under 18 years 
# DP02_0016E Estimate!!HOUSEHOLDS BY TYPE!!Total households!!Average household size 
# DP02_0002E Estimate!!HOUSEHOLDS BY TYPE!!Total households!!Married-couple household 
# DP02_0022E Estimate!!RELATIONSHIP!!Population in households!!Child
# DP02_0024E Estimate!!RELATIONSHIP!!Population in households!!Other nonrelatives
# DP02_0023E Estimate!!RELATIONSHIP!!Population in households!!Other relatives
# DP02_0037E Estimate!!FERTILITY!!Number of women 15 to 50 years old who had a birth in the past 12 months
# DP02_0060E Estimate!!EDUCATIONAL ATTAINMENT!!Population 25 years and over!!Less than 9th grade
# DP02_0061E Estimate!!EDUCATIONAL ATTAINMENT!!Population 25 years and over!!9th to 12th grade, no diploma
# DP02_0062E Estimate!!EDUCATIONAL ATTAINMENT!!Population 25 years and over!!High school graduate (includes equivalency)
# DP02_0065E Estimate!!EDUCATIONAL ATTAINMENT!!Population 25 years and over!!Bachelor's degree
# DP02_0066E Estimate!!EDUCATIONAL ATTAINMENT!!Population 25 years and over!!Graduate or professional degree

pop_evars = [
    'GEO_ID','DP02_0014E','DP02_0016E','DP02_0002E','DP02_0022E','DP02_0024E','DP02_0023E',
'DP02_0037E','DP02_0060E','DP02_0061E','DP02_0062E','DP02_0065E','DP02_0066E'
]

pop_traits = dataframes['population traits'][pop_evars].copy()
pop_traits.columns = pop_traits.iloc[0,:]
pop_traits.drop(pop_traits.index[0],inplace=True)
pop_traits.sort_values(by='Geography',inplace=True)
pop_traits.rename(columns={pop_traits.columns[1]: "<18yrs"},inplace=True)
pop_traits.rename(columns={pop_traits.columns[2]: "avg household size scaled"},inplace=True)
pop_traits.rename(columns={pop_traits.columns[3]: "married household"},inplace=True)
pop_traits.rename(columns={pop_traits.columns[4]: "child"},inplace=True)
pop_traits.rename(columns={pop_traits.columns[5]: "nonrelatives"},inplace=True)
pop_traits.rename(columns={pop_traits.columns[6]: "other relatives"},inplace=True)
pop_traits.rename(columns={pop_traits.columns[7]: "new births"},inplace=True)
pop_traits.rename(columns={pop_traits.columns[8]: "<9th grade"},inplace=True)
pop_traits.rename(columns={pop_traits.columns[9]: "9th-12th grade"},inplace=True)
pop_traits.rename(columns={pop_traits.columns[10]: "HS graduate"},inplace=True)
pop_traits.rename(columns={pop_traits.columns[11]: "college graduate"},inplace=True)
pop_traits.rename(columns={pop_traits.columns[12]: "grad graduate"},inplace=True)

pop_traits['avg household size scaled'] = pd.to_numeric(pop_traits['avg household size scaled'],errors='coerce')

pop_traits['<18yrs scaled'] = pop_traits['<18yrs'].astype(int) / total_pop['total pop'].astype(int)
pop_traits['married household scaled'] = pop_traits['married household'].astype(int) / total_pop['total pop'].astype(int)
pop_traits['child scaled'] = pop_traits['child'].astype(int) / total_pop['total pop'].astype(int)
pop_traits['nonrelatives scaled'] = pop_traits['nonrelatives'].astype(int) / total_pop['total pop'].astype(int)
pop_traits['other relatives scaled'] = pop_traits['other relatives'].astype(int) / total_pop['total pop'].astype(int)
pop_traits['new births scaled'] = pop_traits['new births'].astype(int) / total_pop['total pop'].astype(int)
pop_traits['<9th grade scaled'] = pop_traits['<9th grade'].astype(int) / total_pop['total pop'].astype(int)
pop_traits['9th-12th grade scaled'] = pop_traits['9th-12th grade'].astype(int) / total_pop['total pop'].astype(int)
pop_traits['college graduate scaled'] = pop_traits['college graduate'].astype(int) / total_pop['total pop'].astype(int)
pop_traits['grad graduate scaled'] = pop_traits['grad graduate'].astype(int) / total_pop['total pop'].astype(int)
```

```python
med_income = dataframes['median income'].copy()
med_income = med_income[['GEO_ID','B19013_001E']]
med_income.columns = med_income.iloc[0,:]
med_income.drop(med_income.index[0],inplace=True)
med_income.columns.values[1] = 'median household income'
med_income['median household income scaled'] = pd.to_numeric(med_income['median household income'], errors='coerce')
med_income.dropna(subset=['median household income scaled'],inplace=True)
med_income.isnull().sum()
```

```python
unemployment = dataframes['unemployment'][['GEO_ID','B23025_002E', 'B23025_007E']].copy()
unemployment.columns = unemployment.iloc[0,:]
unemployment.drop(unemployment.index[0],inplace=True)
unemployment.columns.values[1] = 'in labor force'
unemployment.columns.values[2] = 'not in labor force'
# unemployment['in labor force'] = pd.to_numeric(unemployment['in labor force'],errors='coerce')
# unemployment['in labor force'].isnull().any()
unemployment['in labor force scaled'] = unemployment['in labor force'].astype(int) / total_pop['total pop'].astype(int)
unemployment['not in labor force scaled'] = unemployment['not in labor force'].astype(int) / total_pop['total pop'].astype(int) # check up and total_pop in same order
unemployment.dropna(subset=['in labor force scaled','not in labor force scaled'],inplace=True)
unemployment.isnull().sum()
```

```python
# B25034_002E	Estimate!!Total:!!Built 2014 or later	
# B25034_011E	Estimate!!Total:!!Built 1939 or earlier	
# B25034_003E Estimate!!Total:!!Built 2010 to 2013	

home_built_evars = ['GEO_ID','B25034_002E','B25034_011E','B25034_003E']
home_built_year = dataframes['year home built'][home_built_evars].copy()
home_built_year.columns = home_built_year.iloc[0,:]
home_built_year.drop(home_built_year.index[0],inplace=True)
home_built_year.columns.values[1] = '2014 or later'
home_built_year.columns.values[2] = '1939 or earlier'
home_built_year.columns.values[3] = '2010 to 2013'
home_built_year['2014 or later'] = pd.to_numeric(home_built_year['2014 or later'],errors='coerce')
home_built_year['1939 or earlier'] = pd.to_numeric(home_built_year['1939 or earlier'],errors='coerce')
home_built_year['2010 to 2013'] = pd.to_numeric(home_built_year['2010 to 2013'],errors='coerce')
home_built_year.isnull().sum()
```

```python
median_home_val = dataframes['median home value'][['GEO_ID','B25077_001E']].copy()
median_home_val.columns = median_home_val.iloc[0,:]
median_home_val.drop(median_home_val.index[0],inplace=True)
median_home_val.columns.values[1] = 'median home value'
median_home_val['median home value scaled'] = pd.to_numeric(median_home_val['median home value'],errors='coerce')
median_home_val.dropna(subset='median home value scaled',inplace=True) # 1333 - 1310 drops 23 rows
median_home_val.isnull().sum()
```

```python
# B25058_001E median rent
med_rent = dataframes['median rent'][['GEO_ID','B25058_001E']].copy()
med_rent.columns = med_rent.iloc[0,:]
med_rent.drop(med_rent.index[0],inplace=True)
med_rent.columns.values[1] = 'median rent'
med_rent['median rent scaled'] = pd.to_numeric(med_rent['median rent'],errors='coerce')
med_rent['median rent scaled'].isnull().any() # there is null
med_rent.dropna(subset='median rent scaled',inplace=True) #1333 - 1290 = 43 rows dropped
med_rent.isnull().sum()
```

```python
tenure = dataframes['tenture'].copy()
tenure.columns = tenure.iloc[0,:]
tenure.drop(tenure.index[0],inplace=True)
tenure.columns.values[2] = 'total'
tenure.columns.values[4] = 'owner occupied'
tenure.columns.values[6] = 'renter occupied'
tenure['owner occupied scaled'] = tenure['owner occupied'].astype(int)/ tenure['total'].astype(int)
tenure['renter occupied scaled'] = tenure['renter occupied'].astype(int) / tenure['total'].astype(int)
tenure.dropna(subset='owner occupied scaled',inplace=True)
tenure.dropna(subset='renter occupied scaled',inplace=True)
tenure.drop(columns=tenure.columns[8], axis=1,inplace=True)
tenure.isnull().sum()
```

```python
vacant = dataframes['vacant occupied'].copy()
vacant.columns = vacant.iloc[0,:]
vacant.drop(vacant.index[0],inplace=True)
vacant.columns.values[2] = 'total'
vacant.columns.values[4] = 'occupied'
vacant.columns.values[6] = 'vacant'
vacant['occupied_scaled'] = vacant['occupied'].astype(int) / vacant['total'].astype(int)
vacant['vacant_scaled'] = vacant['vacant'].astype(int) / vacant['total'].astype(int)
vacant.dropna(subset='occupied_scaled',inplace=True)
vacant.dropna(subset='vacant_scaled',inplace=True)
vacant.isnull().sum()
```

```python
dfs = [pop_traits, total_pop, med_income, unemployment, home_built_year, med_rent,vacant, tenure]

X_scaled = pd.DataFrame(median_home_val[['Geography','median home value scaled']].copy())

for df in dfs:
    for col in df.columns:
        if isinstance(col, str) and "scaled" in col:
            X_scaled = pd.merge(X_scaled,df[['Geography',col]],on='Geography',how='outer')

X_scaled.columns.values[0]='GEOID'
X_scaled.dropna(inplace=True)
X_scaled.reset_index(inplace=True,drop=True)
X_scaled.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler

first_column = X_scaled.iloc[:, 0]
data_to_standardize = X_scaled.iloc[:, 1:]

scaler = StandardScaler()
standardized_data = scaler.fit_transform(data_to_standardize)

standardized_df = pd.DataFrame(standardized_data, columns=data_to_standardize.columns)
X = pd.concat([first_column, standardized_df], axis=1)
X.to_csv('X.csv',index=False)
```
